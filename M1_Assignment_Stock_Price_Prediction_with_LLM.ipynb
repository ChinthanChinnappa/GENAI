```python
import yfinance as yf
import pandas as pd

# Download historical data for AAPL from 2020-01-01 to 2023-01-01
data = yf.download("AAPL", start="2020-01-01", end="2023-01-01")

# Display summary statistics
print(data.describe())
```

```python
# Calculate moving averages
data['MA_5'] = data['Close'].rolling(window=5).mean()
data['MA_10'] = data['Close'].rolling(window=10).mean()
data['MA_50'] = data['Close'].rolling(window=50).mean()

# Calculate historical volatility (using a 20-day window)
data['Volatility'] = data['Close'].pct_change().rolling(window=20).std() * (252**0.5) # Annualized volatility

# Drop rows with NaN values resulting from the rolling calculations
# Add a check to see if data is empty after dropping NaNs
initial_rows = data.shape[0]
data.dropna(inplace=True)
if data.empty:
    print("Warning: All rows were dropped after adding features and removing NaNs. Please check your data or feature calculations.")
else:
    print(f"Dropped {initial_rows - data.shape[0]} rows with NaN values.")


# Display the first few rows with the new features
display(data.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import pandas as pd # Import pandas

# Create the binary target variable
data['Price_Increase'] = (data['Close'].shift(-1) > data['Close']).astype(int)

# Drop the last row as it will have a NaN for 'Price_Increase'
# Removed the redundant data.dropna(inplace=True) here as it's done after feature engineering


# Define features and target
features = ['Open', 'High', 'Low', 'Close', 'Volume', 'MA_5', 'MA_10', 'MA_50', 'Volatility']
target = 'Price_Increase'

# Ensure data is not empty after feature engineering before proceeding
if not data.empty:
    X = data[features]
    y = data[target]

    # Split data into training and testing sets (70/30 ratio) - Ensure consistency with previous split if possible
    # Using the same random state as before for reproducibility
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Train a Random Forest Classifier model
    model = RandomForestClassifier(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = model.predict(X_test)

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)

    print(f"Accuracy: {accuracy:.2f}")
    print("Confusion Matrix:")
    print(conf_matrix)
    print("Classification Report:")
    print(class_report)

    # Report Feature Importance
    feature_importances = pd.Series(model.feature_importances_, index=features).sort_values(ascending=False)
    print("\nFeature Importance:")
    print(feature_importances)
else:
    print("Data is empty after feature engineering. Cannot proceed with model training.")
```

```python
# Create the binary target variable
data['Price_Increase'] = (data['Close'].shift(-1) > data['Close']).astype(int)

# Calculate moving averages
data['MA_5'] = data['Close'].rolling(window=5).mean()
data['MA_10'] = data['Close'].rolling(window=10).mean()
data['MA_50'] = data['Close'].rolling(window=50).mean()

# Calculate historical volatility (using a 20-day window)
data['Volatility'] = data['Close'].pct_change().rolling(window=20).std() * (252**0.5) # Annualized volatility

# Drop rows with NaN values resulting from the rolling calculations and the shifted target variable
initial_rows = data.shape[0]
data.dropna(inplace=True)

# Add a check to see if data is empty after dropping NaNs
if data.empty:
    print("Warning: All rows were dropped after adding features and removing NaNs. Please check your data or feature calculations.")
else:
    print(f"Dropped {initial_rows - data.shape[0]} rows with NaN values.")


# Display the first few rows with the new features
print("\nFirst 5 rows after feature engineering:")
display(data.head())
print("\nShape after feature engineering:")
print(data.shape)
```
